## Prediksi Tingkat Dampak Bencana di Indonesia Menggunakan Metode Klasifikasi

Dataset:
EM-DAT Indonesia Disaster Profile

Metodologi:
CRISP-DM

# 1. Business Understanding

# 1.1 Latar Belakang

Indonesia merupakan negara yang memiliki tingkat kerentanan tinggi terhadap berbagai jenis bencana alam seperti gempa bumi, banjir, tanah longsor, dan kekeringan.

Data historis bencana dapat dimanfaatkan untuk membangun model prediksi tingkat dampak suatu bencana sehingga dapat membantu proses mitigasi dan pengambilan keputusan.

# 1.2 Rumusan Masalah

Bagaimana membangun model klasifikasi yang dapat memprediksi tingkat dampak suatu bencana berdasarkan karakteristik kejadian bencana?

# 1.3 Tujuan

Bagaimana cara mengelompokkan kejadian bencana berdasarkan karakteristik dampaknya untuk mengidentifikasi pola bencana?

# 1.4 Business Question

- Apakah terdapat kelompok bencana dengan karakteristik yang mirip?
- Cluster mana yang memiliki dampak paling besar?
- Cluster mana yang memiliki dampak paling rendah?
